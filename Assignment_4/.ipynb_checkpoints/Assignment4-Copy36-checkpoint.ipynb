#### Training Images: Describe how you selected the images with respect to the conditions of the detection/recognition methods. 

> The images I selected are shown below and have been included in the attached training images folder. I selected these images firstly by finding individual images of the characters and cropping the areas around each character's face. I cropped these images similar to the examples shown in Lecture 17, where only the individual's face is extracted (i.e. excluding the hair and neck). This was done so that other extraneous features did not interrupt the face recognition process. Additionally, all the images were cropped so that they were the exact same size, which would help with the matrix computations that have to be carried out. The dimensions I chose were of size 138x168. These dimensions were chosen as it was enough to fit the faces of the characters and it is small enough allowing the matrix calculations to be easier and faster to compute. Finally, the images were converted to be in greyscale so that one channel would be dealt with, which also makes the computations much faster and simpler. Specifically, the computations refer to converting the images to eigenspace (and reconstructing the images) which would require more complexity if done across all 3 RGB channels. Thus, to make this easier, the images are dealt with in greyscale and in smaller dimensions.

```python
path1='./train_set/dwight/'

d_mtx = np.empty((30, len(p)))
val = 0
# add all feature maps to the array and then find the average to get the mean feature map
for i in range(1,len(os.listdir(path1))):
    x = cv2.cvtColor(cv2.imread(path1+ str(i)+'.jpg'),cv2.COLOR_BGR2GRAY)
    p = x.flatten()
    d_mtx[val] = p
    val=val+1
        
path2='./train_set/pam/'

# add all feature maps to the array and then find the average to get the mean feature map
for i in range(1,len(os.listdir(path2))):
    x = cv2.cvtColor(cv2.imread(path2+ str(i)+'.jpg'),cv2.COLOR_BGR2GRAY)
    p = x.flatten()
    d_mtx[val] = p
    val=val+1

        
path3='./train_set/jim/'

# add all feature maps to the array and then find the average to get the mean feature map
for i in range(1,len(os.listdir(path3))):
    x = cv2.cvtColor(cv2.imread(path3+ str(i)+'.jpg'),cv2.COLOR_BGR2GRAY)
    p = x.flatten()
    d_mtx[val] = p
    val=val+1
    


mean = np.mean(d_mtx, axis=0)
mean = mean[:, np.newaxis]
Bmean = np.reshape(mean, (168,138))
plt.axis('off')
plt.imshow(np.real(Bmean), cmap="gray")
plt.title('Mean Face Image')
plt.show()

arr1 = np.empty((30, len(p)))
arr1 = d_mtx - mean.T


product = np.dot(arr1, arr1.T)
product = product

eigen_values, eigen_vectors = np.linalg.eig(product)

evals = eigen_values.copy()
eigen_values = eigen_values[:, np.newaxis]

    
prod = np.dot(eigen_vectors, arr1)
eigen_values = np.sqrt((eigen_values), dtype=np.cdouble)
prod = prod/eigen_values

plt.figure(figsize=(15,15))
for c in range(0,5):
    B = np.reshape(prod[c], (168,138))
    plt.subplot(150+c+1)
    plt.imshow(np.real(B), cmap="gray")
    plt.title('Eigenface ' + str(c+1))
    plt.axis('off') 
    
plt.show()
```

```python
def characterRecognition(patch, mean, prod, realChar, listofchar):
    misidentified = False
    recognition = 'Correct'
    patch = patch - mean
    patch10 = np.dot(patch.T,prod.T)
    distarr = []
    
    for x in range(0,30):
        distance = np.linalg.norm(patch10-eigenmatrix[x])
        distarr.append(distance)
    minindx = distarr.index(min(distarr))
    char = name[minindx]
    character = char
    listofchar.append(char)

    neighbourVal = 1
    while char != realChar:
        misidentified = True
        recognition = "Incorrect"
        distarr.pop(minindx)
        minindx = distarr.index(min(distarr))
        char = name[minindx]
        neighbourVal += 1
        
    if misidentified == True:
        title = ('Character misidentified as: ' + character \
              + '\n Number of neighbours needed for \n correct identification: ' + str(neighbourVal))
    else:
        title = ('Correct identification! \n Character identified as: ' + character)
    
    return title, recognition, character, neighbourVal, listofchar
```
