```python
def characterRecognition(patch, mean, prod, realChar, listofchar):
    misidentified = False
    patch = patch - mean
    patch10 = np.dot(patch.T,prod.T)
    distarr = []
    
    for x in range(0,30):
        distance = np.linalg.norm(patch10-eigenmatrix[x])
        distarr.append(distance)
    minindx = distarr.index(min(distarr))
    char = name[minindx]
    character = char
    listofchar.append(char)

    neighbourVal = 1
    while char != realChar:
        misidentified = True
        distarr.pop(minindx)
        minindx = distarr.index(min(distarr))
        char = name[minindx]
        neighbourVal += 1
    return misidentified, character, neighbourVal, listofchar
```
