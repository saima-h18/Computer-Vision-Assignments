### 2. Eigenface Representation

#### Produce and eigenface representation for your training data through PCA. Please note that you are not allowed to use the in-built PCA function in OpenCV/Scikit-Learn. You should implement Snapshot method for PCA (covered in class - Lecture 17) from scratch using numpy. 

```python
# initialize labels array to store the names of the characters
nameLabels = []

# directory path for first character's training images
path1 = './train_set/dwight/'

# initialize D matrix having 30 rows (because 30 training images are used) and 138x168 columns (flattened version
# of the 2D image)
d_mtx = np.empty((30, len(p)))
val = 0
for i in range(1,len(os.listdir(path1))):
    # append the name of character to the labels array
    nameLabels.append("Dwight")
    x = cv2.cvtColor(cv2.imread(path1+ str(i)+'.jpg'),cv2.COLOR_BGR2GRAY)
    p = x.flatten()
    # store the flattened version of the image in the D matrix
    d_mtx[val] = p
    val = val+1

# directory path for second character's training images
path2 = './train_set/pam/'

for i in range(1,len(os.listdir(path2))):
    # append the name of character to the labels array
    nameLabels.append("Pam")
    x = cv2.cvtColor(cv2.imread(path2+ str(i)+'.jpg'),cv2.COLOR_BGR2GRAY)
    p = x.flatten()
    # store the flattened version of the image in the D matrix
    d_mtx[val] = p
    val = val+1

        
# directory path for third character's training images
path3 = './train_set/jim/'

for i in range(1,len(os.listdir(path3))):
    # append the name of character to the labels array
    nameLabels.append("Jim")
    x = cv2.cvtColor(cv2.imread(path3+ str(i)+'.jpg'),cv2.COLOR_BGR2GRAY)
    p = x.flatten()
    # store the flattened version of the image in the D matrix
    d_mtx[val] = p
    val=val+1
    
# find the mean face image of the D matrix
mean = np.mean(d_mtx, axis=0)
mean = mean[:, np.newaxis]
# Plot the mean image
Bmean = np.reshape(mean, (168,138))
plt.axis('off')
plt.imshow(np.real(Bmean), cmap="gray")
plt.title('Mean Face Image')
plt.show()

# find the shifted D matrix by subtracting the mean from the D matrix
arr1 = np.empty((30, len(p)))
arr1 = d_mtx - mean.T

# Snapshot Method:
# Multiply the D matrix and tranposed D matrix such that a smaller dimensioned matrix is outputted
# In this case the product is a 30 x 30 matrix
product = np.dot(arr1, arr1.T)

# Find the eigenvalues and eigenvectors of the product
eigen_values, eigen_vectors = np.linalg.eig(product)

# copy of the eigenvalues to be used in later exercises - the eigenvalues of the larger and smaller matrix
# are the same
evals = eigen_values.copy()
eigen_values = eigen_values[:, np.newaxis]

# to get the eigenvectors of the bigger sized matrix from the eigenvectors of the smaller sized matrix
# multiply the eigenvectors of the small matrix by the D matrix and divide by the square root of the 
# previously found eigenvalues to get the eigenvectors of the larger matrix
prod = np.dot(eigen_vectors, arr1)
eigen_values = np.sqrt((eigen_values), dtype=np.cdouble)
evectors = prod/eigen_values

# Plot the first five eigenfaces using the found eigenvectors
plt.figure(figsize=(15,15))
for c in range(0,5):
    B = np.reshape(evectors[c], (168,138))
    plt.subplot(150+c+1)
    plt.imshow(np.real(B), cmap="gray")
    plt.title('Eigenface ' + str(c+1))
    plt.axis('off') 
plt.show()
```

#### Plot the variance against the eigenvector number in descending order (i.e. total variance vs. number of principal components)

```python
# Sort the eigenvalues in decreasing order
sortedevals = np.sort(evals)[::-1]

tmparr = []
summ = 0

# Sum up the eigenvalues which correspond to the variance
for elem in range(0,30):
    val = sortedevals[elem]
    summ = (summ + val)
    tmparr.append(summ)
# store in a new array
tmparr = np.asarray(tmparr)
# normalize the array by dividing by the max value in the array
maxval = np.max(tmparr)
tmparr = tmparr/maxval

# store the indices of the eigenvector numbers from values 1 to 30
indx = np.arange(1,31)

# Plot the figure
plt.figure(figsize = (10,5))
plt.grid()
plt.plot(indx, tmparr)
plt.title("Fraction of Total Variance as a function of the Eigenvector Number in Descending Order")
plt.xlabel("Eigenvector Index")
plt.ylabel("Fraction of Total Variance")
plt.show()
```

#### Plot the normalized variance (eigenvalues) against the eigenvector number used for computation.

```python
# As can be observed from the previous graph, all the eigenvector values are not needed for computation since 
# 20 eigenvectors from the total 30 correspond to about 95% of the total variance. 

evalsForComputation = sortedevals
indx1 = np.arange(1, 31)
# eigenvalues are normalized by dividing by the max value
normEvals = evalsForComputation/max(evalsForComputation)

# Plot the figure
plt.figure(figsize = (10,5))
plt.grid()
plt.plot(indx1, normEvals)
plt.title("Normalized variance (eigenvalues) as a function of the Eigenvector Number")
plt.xlabel("Eigenvalue Index")
plt.ylabel("Normalized Eigenvalue")
plt.show()
```

### 3. Neighbourhood Clustering

#### For every training image, find the nearest neighbour (L2 distance), and check whether both images belong to the same person. What fraction of your data has a neighbour that is of the same person in high dimensional space?

```python
# use labels as a numpy array
nameLabels = np.asarray(nameLabels)

# initializations
right = 0
accuracy = 0 

# Compute the euclidean distance between each image in high dimensional space with every other image
# and compare the label of the images with the smallest euclidean distance. If it is the same label, it
# is a right match and if not it is incorrect
for x in range(0,30):
    distarr=[]
    for y in range(0,30):
        # ignore the image being compared by itself
        if x==y:
            continue
        # compute euclidean distance
        distance = np.linalg.norm(d_mtx[x]-d_mtx[y])
        # append to the distarr array
        distarr.append(distance)
    # find the minimum euclidean distance
    minindx = distarr.index(min(distarr))
    if nameLabels[x] == nameLabels[minindx]:
        right = right + 1

# find the accuracy
accuracy = (right/30)*100
print('%0.2f %% of the data has a neighbour that is of the same person in high dimensional space.' % accuracy)
```

#### For every training image, find the nearest neighbour (L2 distance), and check whether both images belong to the same person. What fraction of your data has a neighbour that is of the same person in eigenspace?

```python
# Converting all the images to eigenspace:
# The formula on Slide 47 of Lecture 17 was used to derive the eigenspace coordinates of the image
# and to perform the reconstruction
# Extract the first 20 eigenvectors (having the most variance according to the second graph above)
evectors = evectors[0:20,0:]

# initalize an empty 20 x 20 matrix storing the images in eigenspace
eigenmatrix = np.empty((30,20))
for num in range(0,30):
    ex = arr1[num]
    ex = ex[:,np.newaxis]
    ex = ex.T
    dproj = np.dot(ex,evectors.T)
    eigenmatrix[num] = dproj
    
# Perform reconstruction using the images in eigenspace and the eigenvectors along with 
# the mean face image
reconstruct = np.empty((30,138*168))
    
for val in range(0,30):
    facespace = eigenmatrix[val]
    facespace = facespace[:,np.newaxis]
    tmp = facespace*evectors
    tmp = np.sum(tmp, axis=0)
    tmp = tmp[:,np.newaxis].T
    sumval = tmp + mean.T
    reconstruct[val] = sumval
    
# initializations
right = 0
accuracy = 0 
# Compute the euclidean distance between each image in eigenspace with every other image
# and compare the label of the images with the smallest euclidean distance. If it is the same label, it
# is a right match and if not it is incorrect
for x in range(0,30):
    distarr=[]
    for y in range(0,30):
        # ignore the image being compared by itself
        if x==y:
            continue
        # compute euclidean distance
        distance = np.linalg.norm(eigenmatrix[x]-eigenmatrix[y])
        # append to the distarr array
        distarr.append(distance)
    # find the minimum euclidean distance
    minindx = distarr.index(min(distarr))
    if nameLabels[x] == nameLabels[minindx]:
        right = right+1
        
# find the accuracy
accuracy = (right/30)*100
print('%0.2f %% of the data has a neighbour that is of the same person in eigenspace.' % accuracy)
```

#### Compare the accuracy values from high dimensional space and eigenspace. Would you expect there to be a significant difference?

> The accuracy value from the high dimensional space was found to be 66.67% while the accuracy value from the eigenspace was found to be 63.33%. A significant difference was not expected since the eigenface coordinates make use of the 20 eigenvectors which contain the most variation in the data, thus because of the large amount of variation contained in the eigenspace, the accuracy would be expected to be similar to that of the high dimensional space. Additionally, the training images from the high dimensional space are used to create the eigenspace so there should not be a great difference in accuracy levels. This was the expectation and after performing the calculation, this was found to be true as the accuracy values are quite close.

### 4. Face Detection and Recognition

#### You will now detect all the faces in the group image. Use a sliding window to detect the faces. Set a threshold on the distance in eigenspace between the window contents and your training data. Try different values of thresholds and use the one which gives you good results. Display your image with bounding boxes around detected faces for your best threshold.

```python
# directory path for the test image containing all three characters
path4 = './test_image.jpg'

# read image
image = cv2.cvtColor(cv2.imread(path4),cv2.COLOR_BGR2GRAY)
copy = image.copy()

# initializations
list1=[]
list2=[]
counter = 0
pltvar = 1
stridex = 42
stridey = 71
(width, height) = (138,168) # size of window
faceFound = False
plt.figure(figsize=(10,10))

# sliding window across all the pixels according to the determined strides
for y1 in range(0, image.shape[0]-stridey, stridey):
    for x1 in range(0, image.shape[1]-stridex, stridex):
        # creating the window
        window = image[y1:y1+height, x1:x1+width]
        
        # disregard windows that are not the appropriate size
        if window.shape[0]!= 168 or window.shape[1]!=138:
            continue
        
        # flatten the 2D window
        window = window.flatten()
        window = window[:,np.newaxis]

        total = np.zeros(len(window))
        total = total.T
        total = total[:, np.newaxis]
        
        # perform the reconstruction of the window using the formula from Slide 47 Lecture 17
        for num in range(0,20):
            y = window-mean
            mult = np.real(evectors[num])
            mult = mult[:, np.newaxis]
            a1 = np.dot(y.T,mult)
            val=(a1*mult)
            total = total+val
        final = mean + total
        
        distarr = np.empty((30,1),dtype='double')
        # Compute euclidean distance between the window and all the reconstructed training images
        for x in range(0,30):
            distance = np.linalg.norm(final.T-reconstruct[x])
            distarr[x] = (distance)
        # perform normalization
        distarr = distarr / max(distarr)
        # go through all the stored euclidean distances to find the one that is lower than a specific
        # threshold, indicating that a face is present
        for err in distarr:
            if err<0.0415:
                faceFound = True
                
        # if a face is found, display the window and display the bounding boxes in the test image
        if faceFound == True:
            if pltvar == 6:
                plt.axis('off')
#                 plt.show()
                plt.figure(figsize=(10,10))
                pltvar = 1
            # plot the patch with the found coordinates indicating a face is present
            patch = image[y1:y1 + height, x1:x1+width]
            plotval = 150 + pltvar
            plt.axis('off')
            plt.subplot(int(plotval))
            plt.text(x=0,y=0,s= 'Window # ' + str(counter))
            plt.imshow(patch, cmap = 'gray')
            # display bounding box on test image
            copy = cv2.rectangle(copy, (x1, y1), (x1 + width, y1 + height), (255,255,255), 2)
            # keep track of all the x and y coordinates that indicates a face has been found for later sections
            list1.append(x1)
            list2.append(y1)
            faceFound = False
            counter += 1
            pltvar += 1

# display all windows and test image
print('Total of ' + str(counter) + ' detections')
plt.axis('off')
plt.show()
plt.figure(figsize=(10,10))
plt.imshow(copy, cmap='gray')
plt.axis('off')
plt.show()
```

#### You will now try to recognize each of the people in the group image. For the windows in which a face is correctly detected, project the contents of the window to your eigenspace and find the nearest neighbour from your data. The identity of the detected face will be based on the identity of the closest person in the training set.

```python
def characterRecognition(patch, mean, evectors, realChar, listofchar, nameLabels):
    ''' The characterRecognition function takes in the patch or extracted window with the face to be compared, the
        mean face image, the eigenvectors - to perform the eigenspace transformation, the real name of the 
        character in the patch or window and the list of characters array that contains all of the predicted names 
        of the 3 patches. The labels of all the training images is also a parameter of the function.
        
        This function returns information on whether the character in the patch was identified 
        correctly and the amount of neighbours needed to get the correct identity if the character was initially
        misidentified - returned as title in string form. The function also returns the character that the patch 
        is predicted to be through the eigenspace comparison and the number of neighbours needed to get the 
        correct identity as independent values, as well as, the updated list of characters array.
        
        The characterRecognition is done by transforming the extracted window or patch to eigenspace coordinates
        and being compared to all the training images in eigenspace.
    '''
    # Initializations
    misidentified = False
    recognition = 'Correct'
    patch = patch - mean
    patch10 = np.dot(patch.T,evectors.T)
    distarr = []
    neighbourVal = 1
    nameLabels = nameLabels.tolist()

    # Find the euclidean distance between the extracted patch in eigenspace with the training images in eigenspace
    # The label of the training image in eigenspace with the smallest distance is compared to the real character's
    # name which was taken as input
    for x in range(0,30):
        distance = np.linalg.norm(patch10-eigenmatrix[x])
        distarr.append(distance)
    minindx = distarr.index(min(distarr))
    char = nameLabels[minindx]
    character = char
    listofchar.append(char)
    
    # if the characters' names do not match, then the next smallest distance is found and the next pair
    # of indices (with the characters' names) are compared. The number of neighbours (iterations of this
    # process) are kept track of as this happens
    while char != realChar:
        misidentified = True
        recognition = "Incorrect"
        distarr.pop(minindx)
        nameLabels.pop(minindx)
        minindx = distarr.index(min(distarr))
        char = nameLabels[minindx]
        neighbourVal += 1
        
    # If the character is identified correctly or incorrectly, the following titles are returned to be plotted
    # with the matching patches or extracted windows
    if misidentified == True:
        title = ('Character misidentified as: ' + character \
              + '\n Number of neighbours needed for \n correct identification: ' + str(neighbourVal))
    else:
        title = ('Correct identification! \n Character identified as: ' + character)
    
    return title, recognition, character, neighbourVal, listofchar
```

```python
# Initializations

# Window numbers containing a full face 

# Window containing Dwight
win1 = 10
# Window containing Pam
win2 = 16
# Window containing Jim
win3 = 6

# Extracting the patches corresponding to the window numbers for all the characters
patch1 = image[list2[win1]:list2[win1] + height, list1[win1]:list1[win1]+width]
patch2 = image[list2[win2]:list2[win2] + height, list1[win2]:list1[win2]+width] 
patch3 = image[list2[win3]:list2[win3] + height, list1[win3]:list1[win3]+width]
patch1copy = patch1.copy()
patch2copy = patch2.copy()
patch3copy = patch3.copy()
patch1 = patch1.flatten()
patch2 = patch2.flatten()
patch3 = patch3.flatten()
patch1 = np.asarray(patch1)
patch2 = np.asarray(patch2)
patch3 = np.asarray(patch3)
patch1 = patch1[:,np.newaxis]
patch2 = patch2[:,np.newaxis]
patch3 = patch3[:,np.newaxis]

listofchar = []
# Names of all the characters in the extracted patches to be compared to the labels of the training images
realChar = ['Dwight', 'Pam', 'Jim']

plt.figure(figsize=(11,11))

# character recognition being performed for the first character - Dwight
title1, recognition1, character1, neighbourVal1, listofchar = characterRecognition(patch1, mean, evectors, \
                                                                            realChar[0], listofchar, nameLabels)
# Plot the patch with the results from the characterRecognition function
plt.subplot(131)
plt.title(title1)
plt.axis('off')
plt.imshow(patch1copy, cmap='gray')

# character recognition being performed for the second character - Pam
title2, recognition2, character2, neighbourVal2, listofchar = characterRecognition(patch2, mean, evectors, \
                                                                            realChar[1], listofchar, nameLabels)
# Plot the patch with the results from the characterRecognition function
plt.subplot(132)
plt.title(title2)
plt.axis('off')
plt.imshow(patch2copy, cmap='gray')

# character recognition being performed for the third character - Jim
title3, recognition3, character3, neighbourVal3, listofchar = characterRecognition(patch3, mean, evectors, \
                                                                            realChar[2], listofchar, nameLabels)
# Plot the patch with the results from the characterRecognition function
plt.subplot(133)
plt.title(title3)
plt.axis('off')
plt.imshow(patch3copy, cmap='gray')
plt.show()

# Text summary of all character recognitions procedures with all three characters
print('**First Image Summary**\nIdentification: ' + recognition1 + "\nReal Character's Name: " + realChar[0] + \
      '\nRecognized as: ' + character1 + '\nNumber of neighbours needed to get correct label: ' + \
      str(neighbourVal1))
print('\n**Second Image Summary**\nIdentification: ' + recognition2 + "\nReal Character's Name: " + realChar[1]+\
      '\nRecognized as: ' + character2 + '\nNumber of neighbours needed to get correct label: ' + \
      str(neighbourVal2))
print('\n**Third Image Summary**\nIdentification: ' + recognition3 + "\nReal Character's Name: " + realChar[2] + \
      '\nRecognized as: ' + character3 + '\nNumber of neighbours needed to get correct label: ' + \
      str(neighbourVal3))

# Plot the test image with the character recognition predictions
copiedimg = image.copy()
copiedimg = cv2.rectangle(copiedimg, (list1[win1], list2[win1]), (list1[win1] + width, list2[win1] + \
                                                        height), (255,255,255), 2) # bounding box on image
cv2.putText(copiedimg,listofchar[0],((list1[win1], list2[win1]-10)),cv2.FONT_HERSHEY_SIMPLEX,0.7,(255,0,255),2)
copiedimg = cv2.rectangle(copiedimg, (list1[win2], list2[win2]), (list1[win2] + width, list2[win2] + \
                                                        height), (134,0,100), 2) # bounding box on image
cv2.putText(copiedimg,listofchar[1],((list1[win2], list2[win2]-10)),cv2.FONT_HERSHEY_SIMPLEX,0.7,(255,0,255),2)
copiedimg = cv2.rectangle(copiedimg, (list1[win3], list2[win3]), (list1[win3] + width, list2[win3] + \
                                                        height), (255,255,255), 2) # bounding box on image
cv2.putText(copiedimg,listofchar[2],((list1[win3], list2[win3]-10)),cv2.FONT_HERSHEY_SIMPLEX,0.7,(255,255,255),2)
plt.figure(figsize=(13,13))
plt.axis('off')
plt.imshow(copiedimg, cmap = "gray")
plt.show()
```
